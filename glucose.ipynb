# Check if sugar mesurements can be merged together

```python
import pandas as pd
```

```python
dfSerum = pd.read_csv("./tmp/glucose_serum.csv")
dfSerum = dfSerum[dfSerum["valuenum"] < 9000]
dfSerum.describe()
```

```python
axSerum = dfSerum["valuenum"].plot.hist(range=(0,1000), bins=30, color="skyblue", edgecolor="black")
axSerum
```

```python
dfStick = pd.read_csv("tmp/glucose_stick.csv")
dfStick = dfStick[dfStick["valuenum"] < 9000]
dfStick.describe()
```

```python
axStick = dfStick["valuenum"].plot.hist(range=(0,1000), bins=30, color="skyblue", edgecolor="black")
axStick
```

```python
dfWhole = pd.read_csv("tmp/glucose_whole_blood.csv")
dfWhole = dfWhole[dfWhole["valuenum"] < 9000]
dfWhole.describe()
```

```python
axWhole = dfWhole["valuenum"].plot.hist(range=(0,1000), bins=30, color="skyblue", edgecolor="black")
axWhole
```

## Merge glucose mesures 

```python
dfSugar = pd.concat([dfSerum, dfStick, dfWhole])
dfSugar
```

```python
from constants import TEMP_PATH


dfSugar.to_csv(TEMP_PATH / "chartevent_sugar.csv")
```
